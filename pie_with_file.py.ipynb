```python
fname=input("Enter the filename\n")
```

```python
print("Enter the five data in file\n")
with open(fname,"w") as fn: #fn=open(fname,"w+")
    for i in range(0,5):
        x=int(input())
        fn.write(str(x)+'\n')
# 
```

```python
label=[]
with open(fname,"r+") as fr:
        for number in fr:
            label.append(int(number))
print(label)

fr.close()
```
