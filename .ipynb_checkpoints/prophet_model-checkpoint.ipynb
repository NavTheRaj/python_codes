```python
import pandas as pd
import numpy as np
from fbprophet import Prophet
import matplotlib.pyplot as plt

%matplotlib inline
 
plt.rcParams['figure.figsize']=(20,10)
plt.style.use('ggplot')
```

```python
dframe=pd.read_csv("retail_sales.csv", sep=",",parse_dates=True)
```

```python
print(dframe.head())
```

```python
dframe.columns = ['ds','y']
```

```python
dframe.head()
```

```python
dframe.tail()
```

```python
dframe.set_index('ds').y.plot()
```

```python
#NOW LOG TRANSFERRING THE DATA TO CONVERT THE NON STATIONARY TO NON-STATIONARY 
```

```python
dframe['y_orig'] = dframe['y'] # to save a copy of the original data for future use
# log-transforming y
dframe['y'] = np.log(dframe['y'])
```

```python
dframe.head()
```

```python
dframe.set_index('ds').y.plot()
```

```python
#NOW MODELLING THE DATA BY INSTATIATING THE PROPHET OBJECT AND FITTING THE DATA
```

```python
model=Prophet()
model.fit(dframe)
```

```python
future_data = model.make_future_dataframe(periods=24, freq = 'm') # frequency='d' for daily prediction
```

```python
future_data.tail()
```

```python
forecast_data = model.predict(future_data)
```

```python
forecast_data.tail()
```

```python
forecast_data[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head()
```

```python
forecast_data[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
model.plot(forecast_data)
```

```python
dframe.set_index('ds', inplace=True)
forecast_data.set_index('ds', inplace=True)
```

```python
viz_df = dframe.join(forecast_data[['yhat', 'yhat_lower','yhat_upper']], how = 'outer')
```

```python
viz_df.head()  #yhat is modeled data whereas y_orig is our sales data and they are in different scale
```

```python
viz_df['yhat_rescaled'] = np.exp(viz_df['yhat'])
```

```python
viz_df.head() 
```

```python
viz_df[['y_orig', 'yhat_rescaled']].plot()
```

```python
dframe.index = pd.to_datetime(dframe.index) #Assuming we have datetime object
connect_date = dframe.index[-2] #selecting the 2nd to last date
```

```python
mask = (forecast_data.index > connect_date)
predict_df = forecast_data.loc[mask]
```

```python
predict_df.head()
```

```python
viz_df = dframe.join(predict_df[['yhat', 'yhat_lower','yhat_upper']], how = 'outer')
viz_df['yhat_scaled']=np.exp(viz_df['yhat'])
```

```python
viz_df
```

```python
viz_df.tail()
```

```python
fig, ax1 = plt.subplots()
ax1.plot(viz_df.y_orig)
ax1.plot(viz_df.yhat_scaled, color='black', linestyle=':')
ax1.fill_between(viz_df.index, np.exp(viz_df['yhat_upper']), np.exp(viz_df['yhat_lower']), alpha=0.5, color='darkgray')
ax1.set_title('Sales (Orange) vs Sales Forecast (Black)')
ax1.set_ylabel('Dollar Sales')
ax1.set_xlabel('Date')

L=ax1.legend() #legend
L.get_texts()[0].set_text('Actual Sales') #legend text for 1st plot
L.get_texts()[1].set_text('Forecasted Sales') #legend text for 2nd plot
```

```python
#NOW ITS TIME FOR DEPLOYING THE MODEL THAT I CRAETED BY USING PROPHET 
#REQUIREMENTS
# PICKLE LIBRARY FOR PICKLING AI KO MODEL KO ACHAR BANAUN TARIKA
```

```python
import pickle
pkl_path = "Prophet.pkl"
with open(pkl_path, "wb") as f:
    # Pickle the 'Prophet' model using the highest protocol available.
    pickle.dump(model, f)

    
    # save the dataframe
    forecast_data.to_pickle("forecast.pkl")
    print("*** Data Saved ***")
    f.close()
```

```python
# read the Prophet model object
with open("Prophet.pkl", 'rb') as f:
    m = pickle.load(f)

fcast = pd.read_pickle("forecast.pkl")
f.close()
```
