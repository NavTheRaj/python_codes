```python
import numpy as np
from numpy import mean
import matplotlib.pyplot as plt
```

```python
fp=open("population.txt","r")
data=fp.readlines()
fp.close()
date=list()
number=list()
new=list()
```

```python
for lines in data:
    lines=lines.strip()
    new=lines.split(',')
    date.append(new[0])
    number.append(new[1])

 
```

```python
print(number,date)
```

```python
plt.scatter(date,number,color='red')
```
